```python
# 개별 주식이 전체 주식시장의 변동에 반응하는 정도를 나타낸 값이 베타다
# 베타 계산하기
import yfinance as yf
import pandas as pd
import datetime as dt
```

```python
# 베타를 구하는 방법을 알아보기 위해 주식시장에 대한 대용치로 KOSPI 지수를, 개별 주식으로는 전통적 고베타 주인 증권주를 이용한다
#KOSPI코드(^KS11)와 증권주인 키움증권의 티커(039490.KS)를 입력한다.
tickers=["^KS11","039490.KS"]
all_data={}
for ticker in tickers:
    all_data[ticker]=yf.download(ticker,
                                 start="2019-01-01",
                                 end="2024-12-31")
all_data
```

```python
all_data.items()
```

```python
prices=pd.DataFrame({tic:data["Close"] for tic,data in all_data.items()},index=[...])
#{tic:data["Close"] for tic,data in all_data.items()}
#dic={tic:data["Close"] for tic,data in all_data.items()}
#dic
#prices=pd.DataFrame([dic])
#prices = pd.DataFrame({tic: data['Close'] for tic, data in all_data.items()},index="Date")
```

```python
ret=prices.pct_change().dropna()
```
