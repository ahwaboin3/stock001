```python
import yfinance as yf
```

```python
# yfinance 패키지의 download() 함수 내에 티커를 입력하면 주가 정보를 받는다
price=yf.download("AAPL")
price.head()
```

```python
# 진행 과정을 출력하고 싶지 않을 시 progress=False 인자를 추가
# start 인자를 추가하면 데이터 다운로드 기간을 변경 변경할 수 있다.
price=yf.download("AAPL",start="2024-01-01",progress=False)
price.head()
```

```python
# 야후 파이낸스 혹은 여러 벤더의 경우 '티커.국가코드' 형태를 통해 이들을 구분한다.
# 도코 일렉트론의 주가를 받는 법
# 일본의 국가코드는 'T'다
price=yf.download("8035.T",progress=False)
price.head()
```
